# MFCC Training

```python
import pandas as pd
import numpy as np
import h5py
import csv
from sklearn.mixture import GaussianMixture
import pickle
from scipy.stats import norm
import math as mt


print("Starting the testing module for MFCC")
hdf=pd.HDFStore('./features/mfcc/timit.hdf',mode='r')
hdf.keys()
df1 = hdf.get('/timit')

x_test = []
x_label = []
for i in range(len(df1)):
    x_test.append(df1["features"][i])
    x_label.append(df1["labels"][i])
X_test=np.delete(x_test,0,axis=1)
a = ['sil', 'w', 'ih', 's', 'ah', 'ch', 'n', 'ae', 't', 'v', 'r', 'f', 'y', 'uw', 'sh', 'l',
 'b', 'iy', 'd', 'ow', 'm', 'dx', 'k', 'eh', 'aa', 'g', 'dh', 'er', 'jh', 'z', 'ay', 'nan',
 'p', 'ey', 'ng', 'aw', 'hh', 'oy', 'uh', 'th']

n=2
print("For MFCC with energy coefficient:")
while(n<=256):
    work = []
    for i in range(40):
        s = str(n)
        probab = []
        x = len(s)
        while(x<3):
            s = "0"+s
            x=x+1
        fileName = "./models/" + s + "/" + str(a[i]) + "_"+ str(n)+"(Included).sav"
        with open(fileName, 'rb') as f:
            gmm = pickle.load(f)
        y = gmm.score_samples(x_test)
        for j in y:
            probab.append(mt.pow(2,j))
        work.append(probab)
    
    prob=np.zeros(len(df1))
    label=np.zeros(len(df1))
    for i in range(len(work)):
        set = work[i]
        for j in range(len(set)):
            if (set[j] > prob[j]):
                label[j] = int(i)
                prob[j] = set[j]
    
    correct=0
    total=0
    for i in range(len(df1)):
        x = label[i]
        x = int(x)
        if a[x] == x_label[i]:
            correct=correct+1
        elif x_label[i]=="" and a[x] == "nan":
            correct=correct+1
        total=total+1
    print(" Accuracy rate for test vectors with n-components:"+str(n)+" are:",correct/total)
    n=n*2
    
n=2
print("For MFCC without energy coefficient:")
while(n<=256):
    work = []
    for i in range(40):
        s = str(n)
        probab = []
        x = len(s)
        while(x<3):
            s = "0"+s
            x=x+1
        fileName = "./models/" + s + "/" + str(a[i]) + "_"+ str(n)+"(Excluded).sav"
        with open(fileName, 'rb') as f:
            gmm = pickle.load(f)
        y = gmm.score_samples(X_test)
        for j in y:
            probab.append(mt.pow(2,j))
        work.append(probab)
    
    prob=np.zeros(len(df1))
    label=np.zeros(len(df1))
    for i in range(len(work)):
        set = work[i]
        for j in range(len(set)):
            if (set[j] > prob[j]):
                label[j] = int(i)
                prob[j] = set[j]
    
    correct=0
    total=0
    for i in range(len(df1)):
        x = label[i]
        x = int(x)
        if a[x] == x_label[i]:
            correct=correct+1
        elif x_label[i]=="" and a[x] == "nan":
            correct=correct+1
        total=total+1
    print("    Accuracy rate for test vectors with n-components:"+str(n)+" are:",correct/total)
    n=n*2
```
