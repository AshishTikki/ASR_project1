#### note: The modules for MFCC training are designed to be resumable,and for that a Models_Created.txt file has been used
#### The txt file keeps track of how many models have been created and creates to upto n components 256
#### For retraining, the txt file needs to be set as "0 2" and it will be set at "0 512" when it is finished

## The following cell is for MFCC training and creating models with energy coefficient

```python
import pandas as pd
import h5py
import csv
import numpy as np
from sklearn.mixture import GaussianMixture
import pickle


hdf=pd.HDFStore('features/mfcc/timit_train.hdf',mode='r')

hdf.keys()
df1 = hdf.get('/timit')
a = df1["labels"].unique()
i=0
n=0
print(a)
a[31] = "nan"
print(a)
while(True):
    with open("Models_Created.txt") as file:
        lines = []
        for line in file:
            lines.append(line.split())
    file.close()
    for line in lines:
        i = int(line[0])
        n = int(line[1])
    if(n > 256):
        break;
    while(True):
        if(i > 39):
            file1 = open("Models_Created.txt","w")
            s3 = "0 " + str(n*2)
            file1.write(s3)
            file1.close()
            break;
        
        s = str(n)
        x = len(s)
        while(x<3):
            s = "0"+s
            x=x+1
        fileName = "models/"+ s+ "/"+a[i]+"_"+str(n)+"(Included).sav"
        
        fileCsv = "features/mfcc/"+str(a[i])+".csv"
        print(fileCsv)
        with open(fileCsv, 'r') as f:
            reader = csv.reader(f)
            your_list = list(reader)

        X_train = your_list
        
        gmm = GaussianMixture(n_components=n,covariance_type="diag")
        gmm.fit(X_train)
        pickle.dump(gmm, open(fileName, 'wb'))
            
        
        i = i+1
        s2 = str(i) + " " + str(n)
        print(s2)
        file1 = open("Models_Created.txt","w")
        file1.write(s2)
        file1.close()
```

## The following cell is for MFCC training and creating models without energy coefficient

```python
import pandas as pd
import h5py
import csv
import numpy as np
from sklearn.mixture import GaussianMixture
import pickle


hdf=pd.HDFStore('features/mfcc/timit_train.hdf',mode='r')

hdf.keys()
df1 = hdf.get('/timit')
a = df1["labels"].unique()
i=0
n=0
print(a)
a[31] = "nan"
print(a)
while(True):
    with open("Models_Created.txt") as file:
        lines = []
        for line in file:
            lines.append(line.split())
    file.close()
    for line in lines:
        i = int(line[0])
        n = int(line[1])
    if(n > 256):
        break;
    while(True):
        if(i > 39):
            file1 = open("Models_Created.txt","w")
            s3 = "0 " + str(n*2)
            file1.write(s3)
            file1.close()
            break;
            
        s = str(n)
        x = len(s)
        while(x<3):
            s = "0"+s
            x=x+1
        
        
        fileName = "models/"+ s+ "/"+a[i]+"_"+str(n)+"(Excluded).sav"
        
        fileCsv = "features/mfcc/"+str(a[i])+".csv"
        print(fileCsv)
        with open(fileCsv, 'r') as f:
            reader = csv.reader(f)
            your_list = list(reader)

        X_train = your_list
        for row in X_train:
            del row[0]
        
        gmm = GaussianMixture(n_components=n,covariance_type="diag")
        gmm.fit(X_train)
        pickle.dump(gmm, open(fileName, 'wb'))
            
        
        i = i+1
        s2 = str(i) + " " + str(n)
        print(s2)
        file1 = open("Models_Created.txt","w")
        file1.write(s2)
        file1.close()
```

### The following cell is for MFCC_delta training and creating models with energy coefficient
